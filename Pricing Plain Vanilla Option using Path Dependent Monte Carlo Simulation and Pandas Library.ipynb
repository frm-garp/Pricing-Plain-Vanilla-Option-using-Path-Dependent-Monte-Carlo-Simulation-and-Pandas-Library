```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from numpy.random import randn
from numpy import random as rn
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
S0 = 2.521
r = 0.0326
sigma = 1.09
T = 1
```

```python
M = 5000
N = 250
```

```python
h = T/N
data = S0*np.ones((M,N+1))
Z = rn.randn(M,N)
```

```python
# initialize list of lists   
df = pd.DataFrame(Z) 
  
# print dataframe. 
df
```

```python
# initialize list of lists   
df1 = pd.DataFrame(data) 
  
# print dataframe. 
df1 
```

```python
for i in range(0,N):
    df1[i+1] = df1[i]*np.exp((r-sigma**2/2)*h+sigma*np.sqrt(h)*df[i])
df1
```

```python
ST = df1.iloc[:,-1]
```

# Plain Vanilla Call Option

```python
K = 1.251
```

```python
df1['Payoff'] = 0.00000
for i in range(0,M):
    if ST[i] > K:
        df1['Payoff'][i] = (ST[i]-K)
    else:
        df1['Payoff'][i] = 0.0000
df1['PV'] = np.exp(-r*T)*df1['Payoff']
df
```

```python
Call = np.mean(df1['PV'])
print("The value of the plain vanilla call option is:","{:.4f}".format(Call))
```

```python
Error = np.std(df1['PV'])/np.sqrt(M)
print("The stochastic error of the plain vanilla call option is:","{:.3%}".format(Error))
```

```python
Prob = sum(df1['Payoff']>0)/M
print("The probability to exercise the plain vanilla call option is:","{:.2%}".format(Prob))
```

# Plain Vanilla Put Option

```python
K = 1.251
```

```python
df1['Payoff'] = 0.00000
for i in range(0,M):
    if K > ST[i]:
        df1['Payoff'][i] = (K-ST[i])
    else:
        df1['Payoff'][i] = 0.0000
df1['PV'] = np.exp(-r*T)*df1['Payoff']
df1
```

```python
Put = np.mean(df1['PV'])
print("The value of the plain vanilla put option is:","{:.4f}".format(Put))
```

```python
Error1 = np.std(df1['PV'])/np.sqrt(M)
print("The stochastic error of the plain vanilla put option is:","{:.3%}".format(Error1))
```

```python
Prob1 = sum(df1['Payoff']>0)/M
print("The probability to exercise the plain vanilla call option is:","{:.2%}".format(Prob1))
```
